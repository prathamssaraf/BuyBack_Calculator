```python
from annual_report_downloader import process_annual_report
from data_extractor import convert_images_to_csv
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Ask the user to enter the company symbol
company_symbol = input("Enter the company symbol: ")
```

# Extracting Data from Annual Reports

```python
years_to_process = ["2023", "2022", "2021", "2020"]  # Adjust the years as needed

# Initialize a list to store the pages with phrases for each year
pages_with_phrases = []

# Initialize a list to store the image file names for each year
image_file_names = []

# Loop over the years and process the annual reports
for year in years_to_process:
    print(f"Processing annual report for {year}...")
    process_annual_report(company_symbol, year, pages_with_phrases)
    image_file_names.append(f"{company_symbol}_Annual_Report_{year}_Shareholding.PNG")
    print(f"Finished processing annual report for {year}")

# Display all the saved images at the end
num_images = len(image_file_names)
fig, axes = plt.subplots(1, num_images, figsize=(16, 4))

for i in range(num_images):
    image = plt.imread(image_file_names[i])
    axes[i].imshow(image)
    axes[i].set_title(f"Year {years_to_process[i]}")
    axes[i].axis("off")

plt.tight_layout()
plt.show()
```
